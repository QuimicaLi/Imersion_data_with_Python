```python
from google.colab import drive
drive.mount("/content/drive")
import pandas as pd
inmuebles = pd.read_csv("/content/drive/MyDrive/inmuebles_bogota.csv")
inmuebles.info()
```

```python
inmuebles.columns
```

```python
columnas = {"Baños":"Banos", "Área":"Area"}
inmuebles = inmuebles.rename(columns=columnas)
inmuebles.columns
```

```python
inmuebles.info()
```

```python
##La columna valor es str, necesitamos que sea valor float, por lo que vamos a hacer un split de la columna. 
##También quitamos la "," en el precio
valor = inmuebles.Valor.str.split(expand = True)
inmuebles["Moneda"] = valor[0]
inmuebles["Precio"] = valor[1]
inmuebles.sample(3)
inmuebles["Precio"] = inmuebles["Precio"].str.replace(".", "",regex=True)
inmuebles[["Precio", "Barrio"]]
```

```python
##Convertimos el "Precio" en un valor más pequeño con menos ceros
inmuebles["Precio_Millon"] = inmuebles.Precio.astype("float")/1000000
inmuebles.info()
```

```python
inmuebles.describe()
```

```python
##Reducimos decimales para una mejor visualización
pd.set_option("display.precision",2)
pd.set_option("display.float_format", lambda x: "%.2f"% x)
inmuebles.describe()
```

```python
inmuebles["Precio_Millon"].plot.hist(bins=100)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,8))
grafica = sns.histplot(data=inmuebles, x="Precio_Millon", kde=True, hue="Tipo")
grafica.set_title("Distribución del valor de los inmuebles")
plt.xlim(50,2000)
plt.show
```

```python
##Precio m2 por barrio
inmuebles["Valor_m2_Millon"] = inmuebles["Precio_Millon"]/inmuebles["Area"]
inmuebles.head()
```

```python
datos_barrio = inmuebles.groupby("Barrio").sum()

```

```python
datos_barrio["Valor_m2_Barrio"] = datos_barrio["Precio_Millon"]/datos_barrio["Area"]
datos_barrio
```

```python
##Diccionario
m2_barrio = dict(datos_barrio["Valor_m2_Barrio"])
```

```python
##Sustituimos con el nuevo valor creado la columna "Valor_m2_Barrio" así cada inmueble dentro del mismo barrio tendrá el mismo valor por m2
inmuebles["Valor_m2_Barrio"] = inmuebles["Barrio"]
inmuebles["Valor_m2_Barrio"] = inmuebles["Valor_m2_Barrio"].map(m2_barrio)
inmuebles.head()
```

```python
top_barrios = inmuebles["Barrio"].value_counts()[:10].index
```

```python
datos_barrio.reset_index(inplace=True)
datos_barrio
```

```python
datos_barrio.query("Barrio in @top_barrios")
```

```python
plt.figure(figsize=(10,8))
ax = sns.barplot(x="Barrio", y="Valor_m2_Barrio",data=datos_barrio.query("Barrio in @top_barrios"))
ax.tick_params(axis="x", rotation=45)
```

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x="Barrio", y="Valor_m2_Millon",data=inmuebles.query("Barrio in @top_barrios & Valor_m2_Millon < 15"))
ax.tick_params(axis="x", rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x="Barrio", y="Area",data=inmuebles.query("Barrio in @top_barrios & Area < 400"))
ax.tick_params(axis="x", rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x="Barrio", y="Precio_Millon",data=inmuebles.query("Barrio in @top_barrios & Precio_Millon < 4000"))
ax.tick_params(axis="x", rotation=45)
plt.show()
```
